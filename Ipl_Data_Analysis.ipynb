# IPL Data Analysis Project

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv("ipl_comprehensive_dataset.csv")
```

## Display First 5 Rows

```python
df.head()
```

## Dataset Information

```python
df.info
```

## Missing Values

```python
df.isnull().sum()
```

## Team-wise Match Wins

```python
df["winner"].value_counts()
```

```python
df["winner"].value_counts().plot(kind='bar', figsize=(10,5))
plt.title("IPL Match Wins by Team")
plt.show()
```

### Insights

Mumbai Indians and Chennai Super Kings have won the highest number of IPL matches, making them among the most successful teams in the tournament.

## Player of the Match Analysis

```python
df['player_of_match'].value_counts().head(10)
```

```python
df['player_of_match'].value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Players with Most Player of the Match Awards")
plt.show()
```

## Toss Winner Analysis

```python
df['toss_winner'].value_counts()
```

```python
df['toss_winner'].value_counts().plot(kind='bar', figsize=(10,5))
plt.title("Matches Won at Toss by Teams")
plt.show()
```

## Matches Played Per Season

```python
df['season'].value_counts().sort_index()
```

```python
df['season'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title("Matches Played Per Season")
plt.xlabel("season")
plt.ylabel("Number of Matches")
plt.show()
```
